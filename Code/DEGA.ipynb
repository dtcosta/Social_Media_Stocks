```python
def word_count2(str):
    counts = dict()
    words = top_stocks2

    for word in words:
        if word in counts:
            counts[word] += 1
        else:
            counts[word] = 1

    return counts

data = print(word_count2(words_collection2)) 
```

```python
dd2=word_count2(words_collection2)
sorted_dd2=sorted(dd2.items(),key=lambda x: x[1], reverse=True)
sorted_df2=pd.DataFrame(sorted_dd2, columns=['Ticker', 'Frequency'])
ready_df=sorted_df2.replace('\$','',regex=True)
ready_df.head(10)
```
